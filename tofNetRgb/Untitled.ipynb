```python
import numpy as np
import matplotlib.pylab as plt
import cv2
from imutils import paths
```

```python
dirname = '614_LP2'
# path = '/Users/inchanji/Downloads/' + dirname + '/'
path = '/mnt/c/Users/Z440_user1/Downloads/' + dirname + '/'
```

```python
imgpaths = list(paths.list_images(path))
imgs = []
for imgpath in imgpaths:
    imgs.append(cv2.imread(imgpath))
```

```python
imgsize = np.shape(imgs[0])
imgmedian = np.zeros(imgsize)
Nimg = len(imgpaths)
```

```python
print(Nimg, len(imgs), imgsize)
```

```python
for i in range(imgsize[0]):
    for j in range(imgsize[1]):
        r = []
        for n in range(Nimg):
            r.append(imgs[n][i,j,0])
        imgmedian[i,j,0]  = np.median(r)
```

```python
plt.imshow(imgmedian[...,0])
```

```python
cv2.imwrite('/mnt/c/Users/Z440_user1/Downloads/'+ dirname + '/' + dirname + '_median.bmp',imgmedian[...,0])
```
